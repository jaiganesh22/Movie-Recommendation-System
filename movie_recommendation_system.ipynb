```python
genre_table = movie_info_file.pivot_table(index='movie_title',values=movie_info_file.columns[6:])
genre_table
```

```python
genre_dict = {'movie_title':[], 'info':[]}
for x in genre_table.index:
    if x not in genre_dict['movie_title']:
        genre_dict['movie_title'].append(x)
        genre = ''
        for y in genre_table.columns:
            if genre_table.loc[x][y] == 1:
                genre += y + ' '
        genre_dict['info'].append(genre)
genre_dict
```

```python
genre_data = pd.DataFrame(genre_dict)
genre_data
```

```python
#Define Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vector = TfidfVectorizer(stop_words='english')
genre_data['info'] = genre_data['info'].fillna('')
tfidf_matrix = tfidf_vector.fit_transform(genre_data['info'])

#Linear Kernel
from sklearn.metrics.pairwise import linear_kernel
sim_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)

#Indices
indices = pd.Series(genre_data.index, index=genre_data['movie_title']).drop_duplicates()


#Recommender Function
def content_based_recommender(title, sim_scores=sim_matrix):
    idx = indices[title]
    sim_scores = list(enumerate(sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    titles = []
    scores ={'gen_score':[]}
    for x,y in sim_scores:
        titles.append(genre_data.loc[x,'movie_title'])
        scores['gen_score'].append(y)
    ans = pd.DataFrame(index=titles,data = scores)
    ans.index.name = 'movie_title'
    return ans
    
genre_similarity_of_return_of_the_jedi = content_based_recommender('Return of the Jedi (1983)')
```

```python
genre_similarity_of_return_of_the_jedi.head(10)
```

```python
similar_to_returnofthejedi = similar_to_returnofthejedi.merge(genre_similarity_of_return_of_the_jedi,on='movie_title')
```

```python
similar_to_returnofthejedi.sort_values(by=['gen_score','correlation_ratings','correlation_age'],inplace=True,ascending=False)
similar_to_returnofthejedi
```
